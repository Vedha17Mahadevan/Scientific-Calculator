```python
#PROGRAMER OPERATIONS

import tkinter as tk
from tkinter import ttk, messagebox

# Conversion functions
def decimal_to_binary(decimal):
    return bin(decimal)[2:]

def decimal_to_octal(decimal):
    return oct(decimal)[2:]

def decimal_to_hex(decimal):
    return hex(decimal).upper()[2:]

def binary_to_decimal(binary):
    return int(binary, 2)

def octal_to_decimal(octal):
    return int(octal, 8)

def hex_to_decimal(hexadecimal):
    return int(hexadecimal, 16)

# Validation functions
def is_binary(num):
    for i in num:
        if i not in "01":
            return False
    return True

def is_octal(num):
    for i in num:
        if i not in "01234567":
            return False
    return True

def is_decimal(num):
    for i in num:
        if i not in "0123456789":
            return False
    return True

def is_hex(num):
    for i in num:
        if i not in "0123456789ABCDEF":
            return False
    return True

# Base Conversion Functionality
def convert_base():
    input_value = entry.get()
    base_from = combo_from.get()
    base_to = combo_to.get()
    
    if base_from == "Decimal":
        if not is_decimal(input_value):
            messagebox.showerror("Invalid Input", "Please enter a valid decimal number.")
            return
        decimal_value = int(input_value)
    elif base_from == "Binary":
        if not is_binary(input_value):
            messagebox.showerror("Invalid Input", "Please enter a valid binary number.")
            return
        decimal_value = binary_to_decimal(input_value)
    elif base_from == "Octal":
        if not is_octal(input_value):
            messagebox.showerror("Invalid Input", "Please enter a valid octal number.")
            return
        decimal_value = octal_to_decimal(input_value)
    elif base_from == "Hexadecimal":
        if not is_hex(input_value):
            messagebox.showerror("Invalid Input", "Please enter a valid hexadecimal number.")
            return
        decimal_value = hex_to_decimal(input_value)
    
    if base_to == "Decimal":
        result.set(str(decimal_value))
    elif base_to == "Binary":
        result.set(decimal_to_binary(decimal_value))
    elif base_to == "Octal":
        result.set(decimal_to_octal(decimal_value))
    elif base_to == "Hexadecimal":
        result.set(decimal_to_hex(decimal_value))

# Arithmetic Operation Functionality
def perform_operation():
    num1 = entry1.get()
    num2 = entry2.get()
    base = combo_base.get()
    operation = combo_operation.get()

    if base == "Binary":
        if not is_binary(num1) or not is_binary(num2):
            messagebox.showerror("Invalid Input", "Please enter valid binary numbers.")
            return
        num1 = int(num1, 2)
        num2 = int(num2, 2)
    elif base == "Octal":
        if not is_octal(num1) or not is_octal(num2):
            messagebox.showerror("Invalid Input", "Please enter valid octal numbers.")
            return
        num1 = int(num1, 8)
        num2 = int(num2, 8)
    elif base == "Decimal":
        if not is_decimal(num1) or not is_decimal(num2):
            messagebox.showerror("Invalid Input", "Please enter valid decimal numbers.")
            return
        num1 = int(num1)
        num2 = int(num2)
    elif base == "Hexadecimal":
        if not is_hex(num1) or not is_hex(num2):
            messagebox.showerror("Invalid Input", "Please enter valid hexadecimal numbers.")
            return
        num1 = int(num1, 16)
        num2 = int(num2, 16)

    if operation == "Add":
        result = num1 + num2
    elif operation == "Subtract":
        result = num1 - num2

    if base == "Binary":
        result_var.set(bin(result)[2:])
    elif base == "Octal":
        result_var.set(oct(result)[2:])
    elif base == "Decimal":
        result_var.set(str(result))
    elif base == "Hexadecimal":
        result_var.set(hex(result).upper()[2:])

# GUI Setup
root = tk.Tk()
root.title("VSVC Calculator")

tab_control = ttk.Notebook(root)
tab1 = ttk.Frame(tab_control)
tab2 = ttk.Frame(tab_control)
tab_control.add(tab1, text='Base Conversions')
tab_control.add(tab2, text='Arithmetic Operations')

# Base Conversions Tab
tk.Label(tab1, text="Enter number:").grid(column=0, row=0, padx=10, pady=10)
entry = tk.Entry(tab1)
entry.grid(column=1, row=0, padx=10, pady=10)

tk.Label(tab1, text="From base:").grid(column=0, row=1, padx=10, pady=10)
combo_from = ttk.Combobox(tab1, values=["Decimal", "Binary", "Octal", "Hexadecimal"])
combo_from.grid(column=1, row=1, padx=10, pady=10)
combo_from.current(0)

tk.Label(tab1, text="To base:").grid(column=0, row=2, padx=10, pady=10)
combo_to = ttk.Combobox(tab1, values=["Decimal", "Binary", "Octal", "Hexadecimal"])
combo_to.grid(column=1, row=2, padx=10, pady=10)
combo_to.current(1)

result = tk.StringVar()
tk.Label(tab1, text="Result:").grid(column=0, row=3, padx=10, pady=10)
tk.Entry(tab1, textvariable=result, state='readonly').grid(column=1, row=3, padx=10, pady=10)

tk.Button(tab1, text="Convert", command=convert_base).grid(column=1, row=4, padx=10, pady=10)

# Arithmetic Operations Tab
tk.Label(tab2, text="Enter first number:").grid(column=0, row=0, padx=10, pady=10)
entry1 = tk.Entry(tab2)
entry1.grid(column=1, row=0, padx=10, pady=10)

tk.Label(tab2, text="Enter second number:").grid(column=0, row=1, padx=10, pady=10)
entry2 = tk.Entry(tab2)
entry2.grid(column=1, row=1, padx=10, pady=10)

tk.Label(tab2, text="Base:").grid(column=0, row=2, padx=10, pady=10)
combo_base = ttk.Combobox(tab2, values=["Decimal", "Binary", "Octal", "Hexadecimal"])
combo_base.grid(column=1, row=2, padx=10, pady=10)
combo_base.current(0)

tk.Label(tab2, text="Operation:").grid(column=0, row=3, padx=10, pady=10)
combo_operation = ttk.Combobox(tab2, values=["Add", "Subtract"])
combo_operation.grid(column=1, row=3, padx=10, pady=10)
combo_operation.current(0)

result_var = tk.StringVar()
tk.Label(tab2, text="Result:").grid(column=0, row=4, padx=10, pady=10)
tk.Entry(tab2, textvariable=result_var, state='readonly').grid(column=1, row=4, padx=10, pady=10)

tk.Button(tab2, text="Calculate", command=perform_operation).grid(column=1, row=5, padx=10, pady=10)

tab_control.pack(expand=1, fill='both')

root.mainloop()
```

```python

#ALGEBRA OPERATIONS

import tkinter as tk
from tkinter import messagebox, simpledialog
import cmath
import numpy as np

class AlgebraApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Algebra Operations")
        
        self.main_menu()

    def main_menu(self):
        self.clear_frame()
        self.label = tk.Label(self.root, text="Algebra Operations", font=("Helvetica", 16))
        self.label.pack(pady=10)

        self.matrix_button = tk.Button(self.root, text="Matrix Operations", command=self.matrix_operations, width=20)
        self.matrix_button.pack(pady=5)

        self.equation_button = tk.Button(self.root, text="Equations", command=self.equation_operations, width=20)
        self.equation_button.pack(pady=5)

        self.exit_button = tk.Button(self.root, text="Exit", command=self.root.quit, width=20)
        self.exit_button.pack(pady=5)

    def clear_frame(self):
        for widget in self.root.winfo_children():
            widget.destroy()

    def matrix_operations(self):
        self.clear_frame()
        self.label = tk.Label(self.root, text="Matrix Operations", font=("Helvetica", 16))
        self.label.pack(pady=10)

        self.sum_button = tk.Button(self.root, text="Sum of Matrices", command=self.sum_of_matrices, width=30)
        self.sum_button.pack(pady=5)

        self.diff_button = tk.Button(self.root, text="Difference of Matrices", command=self.diff_of_matrices, width=30)
        self.diff_button.pack(pady=5)

        self.mul_button = tk.Button(self.root, text="Matrix Multiplication", command=self.matrix_multiplication, width=30)
        self.mul_button.pack(pady=5)

        self.det_button = tk.Button(self.root, text="Determinant of Matrix", command=self.determinant_of_matrix, width=30)
        self.det_button.pack(pady=5)

        self.trans_button = tk.Button(self.root, text="Transpose of Matrix", command=self.transpose_of_matrix, width=30)
        self.trans_button.pack(pady=5)

        self.inv_button = tk.Button(self.root, text="Inverse of Matrix", command=self.inverse_of_matrix, width=30)
        self.inv_button.pack(pady=5)

        self.trace_button = tk.Button(self.root, text="Trace of Matrix", command=self.trace_of_matrix, width=30)
        self.trace_button.pack(pady=5)

        self.sum_rc_button = tk.Button(self.root, text="Sum of rows and columns", command=self.sum_of_rows_and_columns, width=30)
        self.sum_rc_button.pack(pady=5)

        self.back_button = tk.Button(self.root, text="Back", command=self.main_menu, width=30)
        self.back_button.pack(pady=5)

    def equation_operations(self):
        self.clear_frame()
        self.label = tk.Label(self.root, text="Equation Solver", font=("Helvetica", 16))
        self.label.pack(pady=10)

        self.quadratic_button = tk.Button(self.root, text="Quadratic Equation", command=self.solve_quadratic, width=30)
        self.quadratic_button.pack(pady=5)

        self.cubic_button = tk.Button(self.root, text="Cubic Equation", command=self.solve_cubic, width=30)
        self.cubic_button.pack(pady=5)

        self.back_button = tk.Button(self.root, text="Back", command=self.main_menu, width=30)
        self.back_button.pack(pady=5)

    def sum_of_matrices(self):
        r1, c1, a, b = self.read_matrices("Sum of Matrices")
        if a and b:
            if r1 != len(b) or c1 != len(b[0]):
                messagebox.showerror("Error", "Matrix dimensions do not match for addition.")
                return
            result = [[a[i][j] + b[i][j] for j in range(c1)] for i in range(r1)]
            self.display_matrix(result, "Result Matrix (Sum)")

    def diff_of_matrices(self):
        r1, c1, a, b = self.read_matrices("Difference of Matrices")
        if a and b:
            if r1 != len(b) or c1 != len(b[0]):
                messagebox.showerror("Error", "Matrix dimensions do not match for subtraction.")
                return
            result = [[a[i][j] - b[i][j] for j in range(c1)] for i in range(r1)]
            self.display_matrix(result, "Result Matrix (Difference)")

    def matrix_multiplication(self):
        r1, c1, a, r2, c2, b = self.read_matrices_mul("Matrix Multiplication")
        if a and b:
            if c1 != r2:
                messagebox.showerror("Error", "Matrix dimensions do not allow multiplication.")
                return
            result = [[sum(a[i][k] * b[k][j] for k in range(c1)) for j in range(c2)] for i in range(r1)]
            self.display_matrix(result, "Result Matrix (Product)")

    def determinant_of_matrix(self):
        r1, c1, a = self.read_single_matrix("Determinant of Matrix")
        if a:
            if r1 != c1:
                messagebox.showerror("Error", "Matrix is not square, cannot compute determinant.")
                return
            det = np.linalg.det(a)
            messagebox.showinfo("Determinant", f"Determinant of the Matrix: {det}")

    def transpose_of_matrix(self):
        r1, c1, a = self.read_single_matrix("Transpose of Matrix")
        if a:
            result = [[a[j][i] for j in range(r1)] for i in range(c1)]
            self.display_matrix(result, "Transpose Matrix")

    def inverse_of_matrix(self):
        r1, c1, a = self.read_single_matrix("Inverse of Matrix")
        if a:
            if r1 != c1:
                messagebox.showerror("Error", "Matrix is not square, cannot compute inverse.")
                return
            det = np.linalg.det(a)
            if det == 0:
                messagebox.showerror("Error", "Matrix is singular and doesn't have an inverse.")
                return
            inv = np.linalg.inv(a)
            self.display_matrix(inv, "Inverse Matrix")

    def trace_of_matrix(self):
        r1, c1, a = self.read_single_matrix("Trace of Matrix")
        if a:
            if r1 != c1:
                messagebox.showerror("Error", "Matrix is not square, cannot compute trace.")
                return
            trace = sum(a[i][i] for i in range(r1))
            messagebox.showinfo("Trace", f"Trace of the Matrix: {trace}")

    def sum_of_rows_and_columns(self):
        r1, c1, a = self.read_single_matrix("Sum of Rows and Columns")
        if a:
            row_sums = [sum(row) for row in a]
            col_sums = [sum(a[i][j] for i in range(r1)) for j in range(c1)]
            result = "\n".join([f"Sum of row {i+1}: {row_sums[i]}" for i in range(r1)]) + "\n" + \
                     "\n".join([f"Sum of column {j+1}: {col_sums[j]}" for j in range(c1)])
            messagebox.showinfo("Sum of Rows and Columns", result)

    def solve_quadratic(self):
        a = simpledialog.askfloat("Quadratic Equation", "Enter coefficient of x^2:")
        b = simpledialog.askfloat("Quadratic Equation", "Enter coefficient of x:")
        c = simpledialog.askfloat("Quadratic Equation", "Enter constant:")
        if a is None or b is None or c is None:
            return
        dis = b**2 - 4*a*c
        if dis > 0:
            rt1 = (-b + math.sqrt(dis)) / (2*a)
            rt2 = (-b - math.sqrt(dis)) / (2*a)
            messagebox.showinfo("Roots", f"Roots are Real and Distinct: {round(rt1, 4)} and {round(rt2, 4)}")
        elif dis == 0:
            rt = -b / (2*a)
            messagebox.showinfo("Roots", f"Roots are Real and Equal: {round(rt, 4)} and {round(rt, 4)}")
        else:
            real = -b / (2*a)
            imag = cmath.sqrt(abs(dis)) / (2*a)
            rt1 = complex(real, imag)
            rt2 = complex(real, -imag)
            messagebox.showinfo("Roots", f"Roots are Imaginary: {rt1} and {rt2}")

    def solve_cubic(self):
        a = simpledialog.askfloat("Cubic Equation", "Enter coefficient of x^3:")
        b = simpledialog.askfloat("Cubic Equation", "Enter coefficient of x^2:")
        c = simpledialog.askfloat("Cubic Equation", "Enter coefficient of x:")
        d = simpledialog.askfloat("Cubic Equation", "Enter constant:")
        
        if a is None or b is None or c is None or d is None:
            return
        coefficients = [a, b, c, d]
        roots = np.roots(coefficients)
        messagebox.showinfo("Roots", f"Roots are: {roots}")

    def read_matrices(self, operation):
        r1 = simpledialog.askinteger(operation, "Enter number of rows of matrix A:")
        c1 = simpledialog.askinteger(operation, "Enter number of columns of matrix A:")
        if r1 <= 0 or c1 <= 0:
            messagebox.showerror("Error", "Enter natural numbers for rows and columns.")
            return None, None, None, None
        a = self.input_matrix(r1, c1, "Matrix A")

        r2 = simpledialog.askinteger(operation, "Enter number of rows of matrix B:")
        c2 = simpledialog.askinteger(operation, "Enter number of columns of matrix B:")
        if r2 <= 0 or c2 <= 0:
            messagebox.showerror("Error", "Enter natural numbers for rows and columns.")
            return None, None, None, None
        b = self.input_matrix(r2, c2, "Matrix B")

        return r1, c1, a, b

    def read_matrices_mul(self, operation):
        r1 = simpledialog.askinteger(operation, "Enter number of rows of matrix A:")
        c1 = simpledialog.askinteger(operation, "Enter number of columns of matrix A:")
        if r1 <= 0 or c1 <= 0:
            messagebox.showerror("Error", "Enter natural numbers for rows and columns.")
            return None, None, None, None, None, None
        a = self.input_matrix(r1, c1, "Matrix A")

        r2 = simpledialog.askinteger(operation, "Enter number of rows of matrix B:")
        c2 = simpledialog.askinteger(operation, "Enter number of columns of matrix B:")
        if r2 <= 0 or c2 <= 0:
            messagebox.showerror("Error", "Enter natural numbers for rows and columns.")
            return None, None, None, None, None, None
        b = self.input_matrix(r2, c2, "Matrix B")

        return r1, c1, a, r2, c2, b

    def read_single_matrix(self, operation):
        r1 = simpledialog.askinteger(operation, "Enter number of rows of matrix A:")
        c1 = simpledialog.askinteger(operation, "Enter number of columns of matrix A:")
        if r1 <= 0 or c1 <= 0:
            messagebox.showerror("Error", "Enter natural numbers for rows and columns.")
            return None, None, None
        a = self.input_matrix(r1, c1, "Matrix A")

        return r1, c1, a

    def input_matrix(self, rows, cols, matrix_name):
        matrix = []
        for i in range(rows):
            row = []
            for j in range(cols):
                element = simpledialog.askfloat(matrix_name, f"Enter element ({i+1}, {j+1}):")
                row.append(element)
            matrix.append(row)
        return matrix

    def display_matrix(self, matrix, title):
        result_window = tk.Toplevel(self.root)
        result_window.title(title)
        for row in matrix:
            result_label = tk.Label(result_window, text=str(row))
            result_label.pack()

if __name__ == "__main__":
    root = tk.Tk()
    app = AlgebraApp(root)
    root.mainloop()

```

```python
#SCIENTIC CALCULATOR


import tkinter as tk
from tkinter import simpledialog, messagebox
import math

class ScientificCalculator:
    def __init__(self, root):
        self.root = root
        self.root.title("Scientific Calculator")
        self.memory = 0

        self.display = tk.Entry(root, width=40, borderwidth=5, font=('Arial', 16))
        self.display.grid(row=0, column=0, columnspan=4)

        self.create_buttons()

    def create_buttons(self):
        button_texts = [
            ('sin', 1, 0, 1), ('cos', 1, 1, 1), ('tan', 1, 2, 1), ('csc', 1, 3, 1),
            ('sec', 2, 0, 1), ('cot', 2, 1, 1), ('ln', 2, 2, 1), ('log', 2, 3, 1),
            ('!', 3, 0, 1), ('ceil', 3, 1, 1), ('floor', 3, 2, 1), ('%', 3, 3, 1),
            ('C', 4, 0, 1), ('M', 4, 1, 1), ('Exit', 4, 2, 2)
        ]

        for text, row, col, colspan in button_texts:
            tk.Button(self.root, text=text, width=10, height=2, font=('Arial', 12),
                      command=lambda t=text: self.handle_button(t)).grid(row=row, column=col, columnspan=colspan)

    def handle_button(self, button_text):
        if button_text in ('sin', 'cos', 'tan', 'csc', 'sec', 'cot'):
            self.calculate_trig_function(button_text)
        elif button_text in ('ln', 'log'):
            self.calculate_logarithm(button_text)
        elif button_text == '!':
            self.calculate_factorial()
        elif button_text == 'ceil':
            self.calculate_math_function(math.ceil, "Ceil")
        elif button_text == 'floor':
            self.calculate_math_function(math.floor, "Floor")
        elif button_text == '%':
            self.calculate_modulus()
        elif button_text == 'C':
            self.display.delete(0, tk.END)
        elif button_text == 'M':
            messagebox.showinfo("Memory", f"Memory value: {self.memory}")
        elif button_text == 'Exit':
            self.root.quit()

    def calculate_trig_function(self, func_name):
        angle = self.get_input("Enter angle in degrees:")
        if angle is None:
            return

        func_map = {
            'sin': math.sin,
            'cos': math.cos,
            'tan': self.tan_wrapper,
            'csc': lambda x: 1 / math.sin(x),
            'sec': lambda x: 1 / math.cos(x),
            'cot': lambda x: 1 / math.tan(x)
        }

        radian = math.radians(angle)
        result = func_map[func_name](radian)
        self.display_result(f"{func_name}({angle})", result)

    def tan_wrapper(self, radian):
        angle = math.degrees(radian)
        if (angle - 90) % 180 == 0:
            return "undefined"
        else:
            return math.tan(radian)

    def calculate_logarithm(self, func_name):
        number = self.get_input("Enter number:")
        if number is None:
            return

        if number <= 0:
            messagebox.showinfo(func_name, f"{func_name} is undefined for non-positive numbers")
            return

        func_map = {
            'ln': math.log,
            'log': math.log10
        }

        result = func_map[func_name](number)
        self.display_result(f"{func_name}({number})", result)

    def calculate_factorial(self):
        number = self.get_input("Enter a non-negative integer:", True)
        if number is None or number < 0:
            messagebox.showinfo("Factorial", "Entered number is less than zero")
            return

        result = math.factorial(number)
        self.display_result(f"{number}!", result)

    def calculate_math_function(self, func, name):
        number = self.get_input("Enter a number:")
        if number is None:
            return

        result = func(number)
        self.display_result(f"{name}({number})", result)

    def calculate_modulus(self):
        dividend = self.get_input("Enter dividend:")
        if dividend is None:
            return

        divisor = self.get_input("Enter divisor:")
        if divisor is None:
            return

        result = dividend % divisor
        self.display_result(f"{dividend} % {divisor}", result)

    def get_input(self, prompt, is_integer=False):
        if is_integer:
            return simpledialog.askinteger("Input", prompt)
        else:
            return simpledialog.askfloat("Input", prompt)

    def display_result(self, expression, result):
        self.display.delete(0, tk.END)
        self.display.insert(0, f"{expression} = {result}")

if __name__ == "__main__":
    root = tk.Tk()
    app = ScientificCalculator(root)
    root.mainloop()
```

```python
#STANDARD

import tkinter as tk
from tkinter import ttk, messagebox
import math
import re

class StandardCalculatorApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Standard Calculator")
        self.create_widgets()

    def create_widgets(self):
        # Create a display area
        self.display = tk.Entry(self.root, font=("Arial", 18), bd=10, insertwidth=2, width=14, borderwidth=4)
        self.display.grid(row=0, column=0, columnspan=4)

        # Create buttons for numbers and operations
        buttons = [
            '7', '8', '9', '/', 'C',
            '4', '5', '6', '*', '',
            '1', '2', '3', '-', 'x',
            '0', '.', '=', '+', '+/-'
        ]
        
        row, col = 1, 0
        for button in buttons:
            action = lambda x=button: self.click(x)
            tk.Button(self.root, text=button, padx=20, pady=20, command=action).grid(row=row, column=col)
            col += 1
            if col > 4:
                col = 0
                row += 1
        
        # Additional buttons for specific operations
        tk.Button(self.root, text="Basic Calc", padx=10, pady=20, command=self.basic_calc).grid(row=row, column=0)
        tk.Button(self.root, text="Reciprocal", padx=10, pady=20, command=self.reciprocal).grid(row=row, column=1)
        tk.Button(self.root, text="Negation", padx=10, pady=20, command=self.neg).grid(row=row, column=2)

    def click(self, key):
        current_text = self.display.get()
        if key == 'C':
            self.display.delete(0, tk.END)
        elif key == '=':
            try:
                result = eval(current_text)
                self.display.delete(0, tk.END)
                self.display.insert(tk.END, str(result))
            except Exception as e:
                messagebox.showerror("Error", str(e))
        elif key == '':
            try:
                number = float(current_text)
                result = math.sqrt(number)
                self.display.delete(0, tk.END)
                self.display.insert(tk.END, str(result))
            except Exception as e:
                messagebox.showerror("Error", str(e))
        elif key == 'x':
            try:
                number = float(current_text)
                result = pow(number, 2)
                self.display.delete(0, tk.END)
                self.display.insert(tk.END, str(result))
            except Exception as e:
                messagebox.showerror("Error", str(e))
        elif key == '+/-':
            try:
                number = float(current_text)
                result = -number
                self.display.delete(0, tk.END)
                self.display.insert(tk.END, str(result))
            except Exception as e:
                messagebox.showerror("Error", str(e))
        else:
            self.display.insert(tk.END, key)

    def basic_calc(self):
        expression = self.display.get()
        if expression:
            result = self.calculate(expression)
            self.display.delete(0, tk.END)
            self.display.insert(tk.END, str(result))

    def reciprocal(self):
        try:
            number = float(self.display.get())
            if number == 0:
                messagebox.showerror("Error", "Math error")
            else:
                result = 1 / number
                self.display.delete(0, tk.END)
                self.display.insert(tk.END, str(result))
        except ValueError:
            messagebox.showerror("Error", "Invalid input")

    def neg(self):
        try:
            number = float(self.display.get())
            result = -number
            self.display.delete(0, tk.END)
            self.display.insert(tk.END, str(result))
        except ValueError:
            messagebox.showerror("Error", "Invalid input")

    def calculate(self, expression):
        def operation(num1, num2, operator):
            if operator == '+':
                return num1 + num2
            elif operator == '-':
                return num1 - num2
            elif operator == '*':
                return num1 * num2
            elif operator == '/':
                if num2 == 0:
                    return "Error"
                else:
                    return num1 / num2

        # Tokenize the expression
        tokens = re.findall(r'\d+\.?\d*|[()+\-*/]', expression)

        # Operator precedence
        precedence = {'+': 1, '-': 1, '*': 2, '/': 2}

        def evaluate(tokens):
            operators = []
            operands = []
            for token in tokens:
                if token.isdigit() or '.' in token:
                    operands.append(float(token))
                elif token in '+-*/':
                    while (operators and operators[-1] in precedence and
                           precedence[operators[-1]] >= precedence[token]):
                        num2 = operands.pop()
                        num1 = operands.pop()
                        op = operators.pop()
                        operands.append(operation(num1, num2, op))
                    operators.append(token)
                elif token == '(':
                    operators.append(token)
                elif token == ')':
                    while operators[-1] != '(':
                        num2 = operands.pop()
                        num1 = operands.pop()
                        op = operators.pop()
                        operands.append(operation(num1, num2, op))
                    operators.pop()
            while operators:
                num2 = operands.pop()
                num1 = operands.pop()
                op = operators.pop()
                operands.append(operation(num1, num2, op))
            return operands[0]

        return evaluate(tokens)

if __name__ == "__main__":
    root = tk.Tk()
    app = StandardCalculatorApp(root)
    root.mainloop()
```
